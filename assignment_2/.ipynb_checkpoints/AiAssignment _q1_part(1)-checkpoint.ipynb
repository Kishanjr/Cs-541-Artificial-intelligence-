```python
import numpy as np
import pandas as pd
import csv
from collections import namedtuple
from collections import defaultdict
import math
import time
```

```python
start = time.time()
nyc_taxi_data = pd.read_csv('nyc_taxi_data.csv')
nyc_taxi_data
```

```python
nyc_taxi_data_filtered = nyc_taxi_data[['pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','trip_distance']]
nyc_taxi_data_round = nyc_taxi_data_filtered.round(4)
nyc_taxi_data_round.insert(0, 'nodeDropId', range(168152, 168152 + len(nyc_taxi_data_round)))
nyc_taxi_data_round.insert(0, 'nodePickId', range(0, 168152))
                          
nyc_taxi_data_round
    
```

```python
weighted_edges = nyc_taxi_data_round.values.tolist()

graph_dict={}
graph = defaultdict(list)
for i in weighted_edges:
    pick_cord=(i[2],i[3])
    graph[pick_cord].append([ i[4],i[5],i[6]])


weighted_edges
```

```python
i = 0
lat_long_id = {}
for key in graph:
    values = graph[key]
    if key not in lat_long_id:
        i += 1
        lat_long_id[key] = i
    for value in values:
        lat, long = value[0], value[1]
        if (lat,long) not in lat_long_id:
            i += 1
            lat_long_id[lat, long] = i

list_of_nodeid_lat_long = list(lat_long_id.keys())
list_of_nodeid_nodeId = list(lat_long_id.values())


with open('nodes.csv','w') as out:
    csv_out=csv.writer(out)
    csv_out.writerow(['lat','long'])
    for row in list_of_nodeid_lat_long:
        csv_out.writerow(row)
        
csv_input = pd.read_csv('nodes.csv')
csv_input['nodeid'] = list_of_nodeid_nodeId
csv_input.to_csv('nodes.csv', index=False)
```

```python
node_1_2 = []
node_1_2_dist = []
for source in graph:
    destinations = graph[source]
    source_id = lat_long_id[source]
    for destination in destinations:
        # print(destination[1])
        lat, long = destination[0], destination[1]
        lat, long, dist = destination[0], destination[1],destination[2]
        # print(lat,long)
        destination_id = lat_long_id[lat, long]
        node_1_2.append((source_id, destination_id))
        node_1_2_dist.append([source_id, destination_id,dist])

with open('edges.csv','w') as out:
    csv_out=csv.writer(out)
    csv_out.writerow(['nodeid1',' nodeid2'])
    for row in node_1_2:
        csv_out.writerow(row)
# node_1_2
```

```python
node_1_2_dist
def directedGraph(edges):
    graph_new = defaultdict(dict)
    for u, v, dist in edges:
        graph_new[u][v] = dist
        graph_new[v][u] = dist
    return graph_new

directed_weighted_graph=directedGraph(node_1_2_dist)
directed_weighted_graph
```

```python
start = time.time()
def dijkstra(graph, src):
    queue = [src]
    minDistances = {v: 10000000 for v in graph}
    minDistances[src] = 0
    predecessor = {}

    while queue:
        currentNode = queue.pop(0)
        for neighbor in graph[currentNode]:
            
            newDist = minDistances[currentNode] + graph[currentNode][neighbor]
            
            if newDist < minDistances[neighbor]:
                minDistances[neighbor] = min(newDist, minDistances[neighbor])
                queue.append(neighbor)
                predecessor[neighbor] = currentNode

    return minDistances, predecessor

shortest_path_cost, predecessor = dijkstra(directed_weighted_graph, 1)
# predecessor
```

```python
def UCS(graph, src, dest):
    minDistances, predecessor = dijkstra(graph, src)
   
    
    path = []
    currentNode = dest
    while currentNode != src:
        if currentNode not in predecessor:
            print("Path not reachable")
            break
        else:
            path.insert(0, currentNode)
            currentNode = predecessor[currentNode]
    path.insert(0, src)
    
    if dest in minDistances and minDistances[dest] != float("inf"):
        print('cost is ' + str(minDistances[dest]))
        print('And the path is ' + str(path))
        
UCS(directed_weighted_graph, 1, 10)
end = time.time()
print(end - start)
```
