```python
def ngrams(text, n):
    n_grams = []
    text = text.split()
    text =text[::-1]
    for i in range(len(text)):
        ngram_text = text[i: i + n][::-1]
        ngram_text = " ".join(ngram_text)
        n_grams.append(ngram_text)
    return n_grams

def ngram_words_count_dict_train_test(dataset_lines, n):
    ngram_words_count_dict = {}
    for line in dataset_lines:
        ngram_words_pairs = ngrams(line, n)
        for ngram_words in ngram_words_pairs:
            if ngram_words in ngram_words_count_dict:
                ngram_words_count_dict[ngram_words] +=1
            else:
                ngram_words_count_dict[ngram_words] =1

    return ngram_words_count_dict

def perplexity(test_lines, n, curr_word_dict, prev_word_dict, test_data, smoothing, unique_wrds_train ):
    
    smoothing = smoothing
    
    test_log_likelihood = 0
    for line in test_lines:
        w_list = ngrams(line, n)

        for i in w_list:
            wrd_li = i.split(' ')

            current_words = i
            prev_words = (' '.join(wrd_li[:-1]))

            if smoothing == 'no':

                #using try and except block to avoid keyerror if the words are not present in the dict
                try:
                    current_words_prob = curr_word_dict[current_words]
                except KeyError:
                    current_words_prob = 0

                #using try and except block to avoid keyerror if the words are not present in the dict
                try:
                    prev_words_prob = prev_word_dict[prev_words]
                except KeyError:
                    prev_words_prob = 0
                #using try and except block to avoid division by zero error
                try :
                    if current_words_prob / prev_words_prob != 0:
                        test_log_likelihood += np.log2(current_words_prob / prev_words_prob)   
                except ZeroDivisionError:
                    pass

            else:

                if current_words in curr_word_dict:
                    current_words_prob = curr_word_dict[current_words]
                else: 
                    current_words_prob = 0

                if prev_words in prev_word_dict:
                    prev_words_prob = prev_word_dict[prev_words]
                else: 
                    prev_words_prob = 0
                    
                lambda_val = 1
    
                test_log_likelihood += np.log2((current_words_prob + lambda_val) / (prev_words_prob + (lambda_val * (unique_wrds_train))))     

    test_file_words_count = len(df_data.split())

    PERPLEXITY_ = np.power(2, abs((test_log_likelihood / test_file_words_count)))

    return PERPLEXITY_
    
    
```

```python
import numpy as np

def perplexity_of_traindataset(train_dataset, test_dataset):
    

    train_file = open(train_dataset, "rt")
    train_data = train_file.read()
    train_lines = train_data.split('\n')
    train_words = train_data.split()

    unique_wrds_train = len(set(train_words))

    perplexity_score_relative = []

    perplexity_score_smoothing = []

  

    test_file = open(test_dataset, "rt")
    test_data = test_file.read()
    test_lines = test_data.split('\n')

    for ngram in range(2, 8):

        for smoothing in ['no', 'yes']:

            curr_word_dict = ngram_words_count_dict_train_test(train_lines, ngram)

            prev_word_dict = ngram_words_count_dict_train_test(train_lines, ngram-1)

            ppl = perplexity(test_lines, ngram, curr_word_dict, prev_word_dict, test_data, smoothing, unique_wrds_train)

            if smoothing == 'yes':
                perplexity_score_smoothing.append([ngram, ppl])
            else:
                perplexity_score_relative.append([ngram, ppl])

   
    return perplexity_score_smoothing, perplexity_score_relative

```
